# Optimize and Compile the model

## Tunning the Learning rate

```python
def make_model(learning_rate=0.01):
    base_model = Xception(
        weights='imagenet',
        include_top=False,
        input_shape=(150, 150, 3)
    )

    base_model.trainable = False

    #########################################

    inputs = keras.Input(shape=(150, 150, 3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    outputs = keras.layers.Dense(102)(vectors)
    model = keras.Model(inputs, outputs)

    #########################################

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer,
        loss=loss,
        metrics=['accuracy']
    )

    return model
```

```python
scores = {}

for lr in [0.0001, 0.001, 0.01, 0.1]:
    print(lr)

    model = make_model(learning_rate=lr)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[lr] = history.history

    print()
    print()
```

```python
# Define a list of colors
colors = ['blue', 'orange', 'black', 'green']

for idx, (lr, hist) in enumerate(scores.items()):
    color = colors[idx % len(colors)]  
    plt.plot(hist['accuracy'], label=f'train={lr}', color=color, linestyle='--')
    plt.plot(hist['val_accuracy'], label=f'val={lr}', color=color, linestyle='-')
    
plt.xticks(np.arange(10))

plt.ylim(0.60, 1,0)

# Add labels and title
plt.xlabel("Epochs")  # X-axis label
plt.ylabel("Accuracy")  # Y-axis label
plt.title("Accuracy Over Epochs")  # Graph title

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Best performance: learning_rate = 0.001 (with lr = 0.0001 there are over fitting)

# Experimenting with different sizes of inner layer

```python
def make_model(learning_rate=0.01, size_inner=100):
    base_model = Xception(
        weights='imagenet',
        include_top=False,
        input_shape=(150, 150, 3)
    )

    base_model.trainable = False

    #########################################

    inputs = keras.Input(shape=(150, 150, 3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)

    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)

    outputs = keras.layers.Dense(102)(inner)

    model = keras.Model(inputs, outputs)

    #########################################

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer,
        loss=loss,
        metrics=['accuracy']
    )

    return model
```

```python
learning_rate = 0.001

scores = {}

for size in [10, 100, 1000]:
    print(size)

    model = make_model(learning_rate=learning_rate, size_inner=size)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[size] = history.history

    print()
    print()
```

```python
# Define a list of colors
colors = ['blue', 'orange', 'black', 'green']

for idx, (size, hist) in enumerate(scores.items()):
    color = colors[idx % len(colors)]  
    plt.plot(hist['accuracy'], label=f'train={size}', color=color, linestyle='--')
    plt.plot(hist['val_accuracy'], label=f'val={size}', color=color, linestyle='-')

plt.xticks(np.arange(10))
plt.yticks([0.70, 0.80, 0.90, 1.0])

# Add labels and title
plt.xlabel("Epochs")  # X-axis label
plt.ylabel("Accuracy")  # Y-axis label
plt.title("Accuracy Over Epochs")  # Graph title

# Add legend
plt.legend()

# Show the plot
plt.show()
```

# Regularization and Dropout

```python
def make_model(learning_rate=0.01, size_inner=100, droprate=0.5):
    base_model = Xception(
        weights='imagenet',
        include_top=False,
        input_shape=(150, 150, 3)
    )

    base_model.trainable = False

    #########################################

    inputs = keras.Input(shape=(150, 150, 3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)

    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    drop = keras.layers.Dropout(droprate)(inner)

    outputs = keras.layers.Dense(102)(drop)

    model = keras.Model(inputs, outputs)

    #########################################

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer,
        loss=loss,
        metrics=['accuracy']
    )

    return model
```

```python
# The downside of dropout is that youll need more iterations to learn something. Therefor we change the value from 10 to 30.
learning_rate = 0.001
size = 1000

scores = {}

for droprate in [0.0, 0.2, 0.5, 0.8]:
    print(droprate)

    model = make_model(
        learning_rate=learning_rate,
        size_inner=size,
        droprate=droprate
    )

    history = model.fit(train_ds, epochs=30, validation_data=val_ds)
    scores[droprate] = history.history

    print()
    print()
```

```python
# Define a list of colors
colors = ['blue', 'orange', 'black', 'green']

for idx, (droprate, hist) in enumerate(scores.items()):
    color = colors[idx % len(colors)]  
    plt.plot(hist['accuracy'], label=f'train={droprate}', color=color, linestyle='--')
    plt.plot(hist['val_accuracy'], label=f'val={droprate}', color=color, linestyle='-')

plt.ylim(0.40, 1,0)

# Add labels and title
plt.xlabel("Epochs")  # X-axis label
plt.ylabel("Accuracy")  # Y-axis label
plt.title("Accuracy Over Epochs")  # Graph title

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Best performance: droprate = 0.0 (without dropout)

```python
# Show the best two params separately
hist = scores[0.0]
plt.plot(hist['val_accuracy'], label=0.0)

hist = scores[0.2]
plt.plot(hist['val_accuracy'], label=0.2)

# Add labels and title
plt.xlabel("Epochs")  # X-axis label
plt.ylabel("Validation Accuracy")  # Y-axis label
plt.title("Validation Accuracy Over Epochs")  # Graph title

# Add legend
plt.legend()

# Show the plot
plt.show()
```

# Train a 299299 model and Saving the best model

```python
def make_model(input_size=150, learning_rate=0.01, size_inner=100,
               droprate=0.5):

    base_model = Xception(
        weights='imagenet',
        include_top=False,
        input_shape=(input_size, input_size, 3)
    )

    base_model.trainable = False

    #########################################

    inputs = keras.Input(shape=(input_size, input_size, 3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)

    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    drop = keras.layers.Dropout(droprate)(inner)

    outputs = keras.layers.Dense(102)(drop)

    model = keras.Model(inputs, outputs)

    #########################################

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer,
        loss=loss,
        metrics=['accuracy']
    )

    return model
```

```python
input_size = 299

train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
)

train_ds = train_gen.flow_from_directory(
    '../dataset/train',
    target_size=(input_size, input_size),
    batch_size=32
)

val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)

val_ds = train_gen.flow_from_directory(
    '../dataset/validation',
    target_size=(input_size, input_size),
    batch_size=32,
    shuffle=False
)
```

```python
checkpoint = keras.callbacks.ModelCheckpoint(
    'xception_v1_1_{epoch:02d}_{val_accuracy:.3f}.h5',
    save_best_only=True,
    monitor='val_accuracy',
    mode='max'
)
```

```python
learning_rate = 0.001
size = 1000
droprate = 0.0

model = make_model(
    input_size=input_size,
    learning_rate=learning_rate,
    size_inner=size,
    droprate=droprate
)

history = model.fit(train_ds, epochs=30, validation_data=val_ds,
                   callbacks=[checkpoint])
```
